<h3 style="color:#1F6AE1;">1. Introducción y objetivo del análisis</h3>

En esta primera fase del proyecto se desarrolla un Análisis Exploratorio de Datos (EDA) utilizando el Mushroom Dataset, disponible en https://www.kaggle.com/datasets/uciml/mushroom-classification. Este conjunto de datos contiene información descriptiva sobre diferentes tipos de setas, donde cada instancia representa un hongo caracterizado por múltiples variables categóricas relacionadas con su morfología, color, olor, tipo de láminas y hábitat, entre otras.

La variable objetivo (class) clasifica cada seta como comestible (e) o venenosa (p), lo que permite analizar la distribución de ambas categorías y estudiar posibles patrones diferenciales entre ellas.

El objetivo principal de esta fase es comprender en profundidad la estructura y naturaleza del dataset antes de aplicar cualquier técnica de modelado. Para ello, se persigue:

- Evaluar la cardinalidad de cada atributo.

- Identificar posibles variables constantes o poco informativas.

- Analizar la distribución de la variable objetivo.

- Explorar relaciones preliminares entre atributos y la clase.

Dado que se trata de un dataset completamente categórico, el análisis se centra en el estudio de frecuencias, proporciones y cruces entre variables, permitiendo detectar patrones relevantes que puedan influir posteriormente en el proceso de modelado.

El EDA no solo permite evaluar la calidad y coherencia de los datos, sino también justificar decisiones futuras de preprocesamiento como la codificación de variables, la eliminación de atributos irrelevantes o el tratamiento de valores faltantes.

En definitiva, esta fase constituye una etapa fundamental para comprender el comportamiento de las variables y sentar una base sólida para las técnicas de aprendizaje automático que se aplicarán en etapas posteriores del proyecto.

<h3 style="color:#1F6AE1;">3. Análisis de las variables</h3>

Nuestra columna objetivo es clase por lo que es importante saber que variables pueden tener mayor influencia o te da un detalle más especfícico sobre si una seta va a ser venenosa o comestible.

Para esto se ha realizado un análisis de cada variable con respecto a la clase, para así identificar posibles patrones o relaciones que puedan ser útiles para la clasificación.

<h5 style="color:#1F6AE1;">¿Qué variables tienen una cardinalidad muy alta?</h5>

Es importante identificar variables con alta cardinalidad, ya que pueden generar problemas en el modelado, especialmente si se utilizan técnicas de codificación como One-Hot Encoding, que pueden aumentar significativamente el número de características.

En este análisis se han identificado las variables con alta cardinalidad, lo que nos permitirá tomar decisiones informadas sobre cómo manejarlas en el proceso de preprocesamiento y modelado.

<h5 style="color:#1F6AE1;">¿Qué variables tienen una cardinalidad muy baja?</h5>

Las variables con baja cardinalidad pueden ser menos informativas para el modelo, ya que no aportan mucha variabilidad. Identificar estas variables nos ayudará a decidir si es necesario eliminarlas o combinarlas con otras para mejorar la calidad del modelo.

```python
sns.set_style('whitegrid')

cols = [c for c in df_copy.columns if c not in ['class']]

color_map = {'e': '#d62728',  # comestible = rojo
             'p': '#2ca02c'}  # venenoso = verde

for col in cols:
    ct = pd.crosstab(df_copy[col], df_copy['class']).sort_index()
    if ct.shape[0] == 0:
        continue

    ct_pct = ct.div(ct.sum(axis=1), axis=0)

    class_cols = [c for c in ['e','p'] if c in ct.columns]
    colors = [color_map[c] for c in class_cols]

    fig, (ax1, ax2) = plt.subplots(nrows=1, ncols=2, figsize=(14,4))

    # --- Proporciones ---
    ct_pct[class_cols].plot(kind='bar', stacked=True, ax=ax1, color=colors)
    ax1.set_ylabel('Proporción')
    ax1.set_xlabel(col)
    ax1.set_title(f'Proporción comestible vs venenoso por {col}')
    ax1.tick_params(axis='x', rotation=45)
    ax1.legend(['Comestible', 'Venenoso'])

    # --- Conteos ---
    ct[class_cols].plot(kind='bar', ax=ax2, color=colors)
    ax2.set_ylabel('Conteo')
    ax2.set_xlabel(col)
    ax2.set_title(f'Conteo comestible vs venenoso por {col}')
    ax2.tick_params(axis='x', rotation=45)
    ax2.legend(['Comestible', 'Venenoso'])

    plt.tight_layout()
    plt.show()
```

<h3 style="color:#1F6AE1;">Conclusiones del EDA</h3>

- **Distribución de la clase:** el dataset está aproximadamente balanceado (~52% comestibles, ~48% venenosos).

- **Variables con mayor discriminación:** `odor`, `gill-color`, `ring-type` y `stalk-color` (arriba/abajo) muestran asociaciones fuertes con la clase y deben priorizarse al preparar modelos.

- **Soporte y cautela:** algunas categorías presentan asociación perfecta pero con soporte muy bajo (p. ej. 4–16 observaciones). Estas asociaciones pueden ser espurias; tratarlas con prudencia.

- **Recomendaciones de preprocesado:** evitar expansión masiva con One‑Hot en variables de alta cardinalidad; considerar codificaciones alternativas (target encoding, embeddings) y agrupar categorías raras por frecuencia mínima.
